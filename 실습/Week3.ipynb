```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel("x axis label")
plt.ylabel("y axis label")
plt.title("Sine and Cosine")
plt.legend(['Sine', 'Cosine'])
plt.show()
```
